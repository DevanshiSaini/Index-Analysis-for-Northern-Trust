```python
crypto_1 = pd.read_json('https://api.coinmarketcap.com/v1/ticker/?limit=0')

crypto_1
```

```python
crypto_2 = crypto_1[['id', 'market_cap_usd']]
crypto_2.head()
```

```python
crypto_2.count()
```

```python
crypto_3 = crypto_2.query('market_cap_usd > 0')

crypto_3.count()
```

```python
crypto_3
```

```python
crypto_3=crypto_3.set_index('id')


crypto_3.head()
```

```python
market_cap_total=crypto_3['market_cap_usd'].sum()

market_cap_total
```

```python
crypto_3['market_cap_%']=(crypto_3['market_cap_usd']/market_cap_total)*100

crypto_3
```

```python
crypto_3['market_cap_%']
```

```python
import matplotlib.pyplot as plt
plt.plot(crypto_3['market_cap_%'])
```
