### Sort the data in descending order by 'Monthly Return' ; extract the top 5 stocks ; and serialise accordingly

```python
def topstocks(x):
    
    by_performance=x.sort_values(by='Monthly Return', ascending=False)
    top_stocks=by_performance.iloc[:5,[1,5,9]]
    top_stocks_1=top_stocks.set_index('Security Name')
    return top_stocks_1
```
